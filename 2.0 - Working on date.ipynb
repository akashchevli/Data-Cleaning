# Convert our date columns to datetime

Now this is a time to convert out date column datatype into datetime, there is a method for convet column in a datatime, 
to_datetime()

```python
landslide['NewDate'] = pd.to_datetime(landslide['date'])
landslide['NewDate'].head()
```

Yup, we convert date column datatype into datetime. 

# Select just the day of the month from our column

```python
landslide['NewDate'].dt.day
```

```python
# Similar to day we can also select month and year
```

```python
# Let's create date plot
```

```python
sns.distplot(landslide['NewDate'].dt.day, bins=30, kde=False)
```

```python
landslide['NewDate'].value_counts().plot.line()
```

Load another dataset

```python
earthquakes = pd.read_csv('databaseE.csv')
```

```python
earthquakes.head()
```

Check the datatype of Date column

```python
earthquakes['Date'].dtypes
```

Again datatype of Date column is object, So we need to convert that datatype into DateTime

```python
earthquakes['NewDate'] = earthquakes['Date'].apply(pd.to_datetime, utc=True)
```

```python
earthquakes['NewDate']
```

Yeah, We done with datatype, But it also give timestamp, we only need date. so for that we use date attribute 

```python
earthquakes['ChangeDate'] = earthquakes['NewDate'].dt.date
```

We achieve our goal, now it's time to look at the dataset

```python
earthquakes.head()
```

We don't need to require NewDate Column, So we drop it

```python
earthquakes.drop(columns=['NewDate'], inplace=True)
```

```python
earthquakes
```

Succesfully drop the column, inplace=True is allows us to drop the column in original dataset if we don't provide then it will not drop column into the orginal dataset.
